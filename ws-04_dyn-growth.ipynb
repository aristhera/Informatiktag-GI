```python
def main():
    inf = 1
    pop = 8 * 10**5
    print(r_wert(inf, pop))

def r_wert(a, b): # a = inf, b = pop, c = r-wert
    day = 1
    c = 1.1
    while a < b:
        print(day, a, c)
        if len(str(a)) == 7:
            c = 1.1
        elif len(str(a)) == 8:
            c = 1.3
        elif len(str(a)) == 9:
            c = 1.5
        a *= c
        a = round(a, 2)
        day += 1
    return day, a, c

if __name__ == "__main__":
    main()
```
